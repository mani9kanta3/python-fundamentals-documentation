```python
# Create and Print a Dictionary
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

print(student)
```

```python
# Access a Value Using Key
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

print(student["name"])
```

```python
# Add a New Key-Value Pair
student = {
    "name": "Mani",
    "age": 22
}

student["course"] = "Python"

print(student)
```

```python
# Update an Existing Value
student = {
    "name": "Mani",
    "age": 22
}

student["age"] = 23

print(student)
```

```python
# Delete a Key-Value Pair
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

del student["course"]

print(student)
```

```python
# Check if a Key Exists
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

key = input("Enter key: ")

if key in student:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
# Print all keys
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

print(student.keys())
```

```python
# Print all values
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

print(student.values())
```

```python
# Print all key-value pairs
student = {
    "name": "Mani",
    "age": 22,
    "course": "Python"
}

for key, value in student.items():
    print(key, ":", value)
```

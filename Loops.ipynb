```python
# Print numbers from 1 to 10 using a for loop.
for i in range(1, 11):
    print(i)
```

```python
# Print numbers from 10 to 1 using a while loop.
i = 10

while i >= 1:
    print(i)
    i -= 1
```

```python
# Print all even numbers from 1 to 20.
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```python
# Print the sum of the first 10 natural numbers.
total = 0

for i in range(1, 11):
    total += i

print(total)
```

```python
# Print the multiplication table of a given number.
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# Find the factorial of a given number using a while loop.
num = int(input("Enter a number: "))

fact = 1

while num > 0:
    fact *= num
    num -= 1

print("Factorial =", fact)
```

```python
# Check if a number is prime or not.
num = int(input("Enter a number: "))

is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if num > 1 and is_prime:
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
# Find the sum of digits of a given number.
num = int(input("Enter a number: "))

total = 0

while num > 0:
    digit = num % 10
    total += digit
    num //= 10

print(total)
```

```python
# Print all odd numbers from 1 to 50.
for i in range(1, 51):
    if i % 2 != 0:
        print(i)
```

```python
# Find the largest number in a list using a loop.
numbers = [12, 45, 67, 23, 89, 34]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest Number:", largest)
```

```python
# Print all numbers from 1 to 100 that are divisible by 5.
for i in range(1, 101):
    if i % 5 == 0:
        print(i)
```

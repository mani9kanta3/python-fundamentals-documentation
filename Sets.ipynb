```python
# Create a Set and Print It
numbers = {10, 20, 30, 40}

print(numbers)
```

```python
# Add an Element to a Set using add()
numbers = {10, 20, 30}

numbers.add(40)

print(numbers)
```

```python
# Add Multiple Elements using update()
numbers = {10, 20, 30}

numbers.update([40, 50, 60])

print(numbers)
```

```python
# Remove an Element using remove()
numbers = {10, 20, 30, 40}

numbers.remove(30)

print(numbers)
```

```python
# Remove an Element using discard()
numbers = {10, 20, 30, 40}

numbers.discard(30)

print(numbers)
```

```python
# Remove and Return a Random Element using pop()
numbers = {10, 20, 30, 40}

removed = numbers.pop()

print("Removed:", removed)
print(numbers)
```

```python
# Clear All Elements from a Set
numbers = {10, 20, 30}

numbers.clear()

print(numbers)
```

```python
# Find the Length of a Set
numbers = {10, 20, 30, 40}

print(len(numbers))
```

```python
# Check if an Element Exists in a Set
numbers = {10, 20, 30, 40}

num = int(input("Enter a number: "))

if num in numbers:
    print("Found")
else:
    print("Not Found")
```

```python
# Iterate Through a Set using a Loop
numbers = {10, 20, 30, 40}

for num in numbers:
    print(num)
```

```python
# Create a Set from a List to Remove Duplicates
numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = set(numbers)

print(unique_numbers)
```

```python
# Find the Union of Two Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.union(set2))
```

```python
# Find the Intersection of Two Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.intersection(set2))
```

```python
# Find the Difference of Two Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.difference(set2))
```

```python
# Find the Symmetric Difference of Two Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.symmetric_difference(set2))
```

```python
# Check if One Set is a Subset of Another
set1 = {1, 2}
set2 = {1, 2, 3, 4}

print(set1.issubset(set2))
```

```python
# Check if One Set is a Superset of Another
set1 = {1, 2, 3, 4}
set2 = {1, 2}

print(set1.issuperset(set2))
```

```python
# Check if Two Sets Have No Common Elements
set1 = {1, 2}
set2 = {3, 4}

print(set1.isdisjoint(set2))
```

```python
# Copy a Set
numbers = {10, 20, 30}

new_set = numbers.copy()

print(new_set)
```

```python
# Find the Maximum Value in a Set
numbers = {10, 20, 50, 30}

print(max(numbers))
```

```python
# Find the Minimum Value in a Set
numbers = {10, 20, 50, 30}

print(min(numbers))
```

```python
# Find the Sum of All Elements in a Set
numbers = {10, 20, 30, 40}

print(sum(numbers))
```

```python
# Update a Set with Another Set
set1 = {1, 2, 3}
set2 = {4, 5, 6}

set1.update(set2)

print(set1)
```

```python
# Remove All Elements of Another Set from a Set
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4}

set1.difference_update(set2)

print(set1)
```

```python
# Keep Only Common Elements using intersection_update()
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

set1.intersection_update(set2)

print(set1)
```

```python
# Delete an Entire Set
numbers = {10, 20, 30}

del numbers

print("Set deleted successfully")
```

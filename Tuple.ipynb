```python
# Create a tuple containing integers, strings, and floats. Print the tuple.
t = (10, "Python", 3.14)
print(t)
```

```python
# Print the first and last elements of a tuple.
t = (10, 20, 30, 40, 50)

print(t[0])
print(t[-1])
```

```python
# Slice a tuple and print elements from index 1 to 3.
t = (1, 2, 3, 4, 5)

print(t[1:4])
```

```python
# Access the second element of the nested tuple.
t = (1, 2, (3, 4), 5)

print(t[2][1])
```

```python
# Find how many times a specific element appears in a tuple.
t = (1, 2, 3, 1, 4, 1)

print(t.count(1))
```

```python
# Find the index of a specific element in a tuple.
t = ("apple", "banana", "cherry")

print(t.index("banana"))
```

```python
# Concatenate two tuples and print the result.
t1 = (1, 2, 3)
t2 = (4, 5, 6)

print(t1 + t2)
```

```python
# Multiply a tuple by an integer.
t = (1, 2)

print(t * 3)
```

```python
# Find the length of a tuple.
t = (10, 20, 30, 40, 50)

print(len(t))
```

```python
# Check if an element exists in a tuple.
t = (10, 20, 30, 40, 50)

element = int(input("Enter element: "))

if element in t:
    print("Element exists")
else:
    print("Element does not exist")
```

```python
# Delete a tuple and print a message.
t = (10, 20, 30)

del t

print("Tuple deleted successfully")
```
